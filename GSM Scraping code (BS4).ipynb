```python
#importing all the required libraries
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import pandas as pd
```

## GSM Galaxy S8

```python
records=[]
for x in range(1,300):
    page=x
    review_pages=["https://www.gsmarena.com/samsung_galaxy_s8-reviews-8161p" + str(page) + '.php']
    print("Scraping page"+ str(page))
    for url in review_pages:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        for caption in soup.find_all(class_='user-thread'):
            reviews = caption.find('p', attrs={'class':'uopin'}).text
            name = caption.find('li', attrs={'class': ['uname2','uname']}).text
            date = caption.find('li', attrs={'class': 'upost'}).text
            records.append((name, date,reviews))
```

```python
len(records)
```

```python
df1=pd.DataFrame(records, columns=['reviewer_name', 'review_date', 'review_content'])
```

```python
df1.to_csv('GSM_scrape_reviews_samsung_GalaxyS8.csv', index=False, encoding='utf-8')
```
